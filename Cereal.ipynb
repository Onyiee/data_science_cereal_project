```python
import pandas as pd



cereal = pd.read_csv("cereal.csv")
cereal
```

```python
carbohydrate_cereals = {"name": cereal["name"],
                        "sugars": cereal["sugars"],
                        "carbo": cereal["carbo"],
                        "fiber": cereal["fiber"]
}

# cereal[]

```

```python
carbohydrate_cereals_analysis = pd.DataFrame(carbohydrate_cereals, columns = ['name','sugars','fiber', 'carbo'])
```

```python
carbohydrate_cereals_analysis.plot( x ='name', y = ['sugars','fiber', 'carbo'], kind='bar', figsize= (20, 20))
```

```python
carbohydrate_cereals_analysis.plot( x ='name', y = ['sugars','fiber', 'carbo'], figsize= (20, 10), )
```

```python
carbohydrate_cereals_analysis
```

```python
carbohydrate_cereals_analysis_total = carbohydrate_cereals_analysis['sugars'] + carbohydrate_cereals_analysis['fiber'] + carbohydrate_cereals_analysis['carbo']
```

```python
carbohydrate_cereals_analysis_totals = pd.DataFrame(carbohydrate_cereals_analysis_total, columns=["carbohydrate_total"])
```

```python
carbohydrate_cereals_analysis_totals
```

```python
carbohydrate_cereals_analysis_total_enw = {"name": cereal["name"],
                                           "carbohydrate_total": carbohydrate_cereals_analysis_totals["carbohydrate_total"]
                                       }
```

```python
carbohydrate_cereals_analysis_total_enw = pd.DataFrame(carbohydrate_cereals_analysis_total_enw, columns=['name', 'carbohydrate_total'])
```

```python
carbohydrate_cereals_analysis_total_enw
```

```python
carbohydrate_cereals_analysis_total_enw.plot(x ='name', y = 'carbohydrate_total', figsize= (20, 10) , kind="bar")
```

```python
mineral_cereals = {"name": cereal["name"],
                        "potass": cereal["potass"],
                        "sodium": cereal["sodium"]
                      
}
mineral_cereals
```

```python
mineral_cereals_analysis = pd.DataFrame(mineral_cereals, columns = ['name','potass','sodium'])
mineral_cereals_analysis
```

```python
mineral_cereals_analysis.plot( x ='name', y = ['potass','sodium'], kind='bar', figsize= (20, 20))
```

```python
mineral = mineral_cereals_analysis['potass'] + mineral_cereals_analysis['sodium'] 
```

```python
mineral = pd.DataFrame(mineral, columns=[ "mineral"])
mineral = {"name" : mineral_cereals_analysis['name'], 
           "mineral" : mineral['mineral']
           }
minerals = pd.DataFrame(mineral, columns=[ "name","mineral"])
minerals
```

```python

minerals.plot( x ='name', y = 'mineral', kind='bar', figsize= (20, 10), color = 'black')
```

```python
fat = {"name": cereal["name"],
       "fat": cereal["fat"]
                       
}
fat
```

```python
fat_analysis = pd.DataFrame(fat, columns = ['name','fat'])
fat_analysis
```

```python

fat_analysis.plot( x ='name', y = 'fat', kind='bar', figsize= (20, 10), color = 'red')
```

```python
protein = {"name": cereal["name"],
       "protein": cereal["protein"]
                       
}
protein
```

```python
protein_analysis = pd.DataFrame(protein, columns = ['name','protein'])
protein_analysis
```

```python
protein_analysis.plot( x ='name', y = 'protein', kind='bar', figsize= (20, 10), color = 'purple')
```

```python
vitamins = {"name": cereal["name"],
       "vitamins": cereal["vitamins"]
                       
}
vitamins
```

```python
vitamins_analysis = pd.DataFrame(vitamins, columns = ['name','vitamins'])
vitamins_analysis
```

```python
vitamins_analysis.plot( x ='name', y = 'vitamins', kind='bar', figsize= (20, 10), color = 'green')
```

```python
cereal_classes = {
    "name" : cereal['name'],
    "carbohydrate_total" : carbohydrate_cereals_analysis_totals["carbohydrate_total"],
    "mineral" : minerals["mineral"],
    "vitamins" : vitamins_analysis["vitamins"],
    "protein" : protein_analysis["protein"],
    "fat" : fat_analysis["fat"]
}
cereal_classes
```

```python
cereals_table = pd.DataFrame(cereal_classes, columns = ['name','carbohydrate_total','protein','mineral','vitamins','fat'], )
cereals_table
```

```python
cereals_table.plot(x = 'name', y =['carbohydrate_total','protein','fat','mineral','vitamins'],  kind='bar', figsize= (20, 10), color = 'green')
```
